or figure level function like catplot() as in "categorical plot" `

### We can use hue, style and size in categorical Scatterplots too. 

In categorical Scatterplots you use:

`Categorical Data -> X-axis
Numerical Data -> Y-axis`

```python
plt.style.use('fivethirtyeight')
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.catplot(x='day', y="tip", kind='strip', data=tips) #StripPlot
# Due to jitter(which adds noise), if data point have same value, then also it spreads abit.
```

```python
sns.catplot(x='day', y='tip', kind='swarm', data=tips) #SwarmPlot
# sns.swarmplot(x=tips['day'], y=tips['tip'])
# Here the same value is spread side ways rather than near each other as shown below: 
```

# BoxPlots

Box plots are standard way of displaying the distribution of data based on a five number summary minimum , first quartile (Q1), Median, third quartile (Q3) and maximum. It can tell you about your outliers and what their values are. It can also tell you if you data is symmetrical, how tightly your data is grouped and how your data is skewed. 

IQR = Q3- Q1




```python
# Univariate Analysis
sns.boxplot(x='tip', data=data)
```

```python
# Multivariate Analysis

sns.catplot(x='day', y='tip', kind="box", data=data)
```

```python
#Using Hue in this multivariate analysis

sns.catplot(x='day', y='total_bill', kind='box', hue="sex", data=data)
```

# VIOLIN PLOT

Violin plots are the method of plotting numerical data and can be considered a combination of the box with a kernel density plot.

![](https://images.ctfassets.net/fi0zmnwlsnja/sdfgtcRp16wTNOcRceGQm/5bfcb73d2261d49ff20dd7857e0152b1/Screen_Shot_2019-03-01_at_11.36.10_AM.png)

The unquestionable advantage of the violin plot over box plot is that aside from showing the above mentioned statistics it also shows the entire distribution of the data.

```python
sns.violinplot(x=tips['tip'])
```

It can also be used to draw violinplot among different categories

```python
sns.catplot(x='day', y='total_bill', kind='violin', hue='sex', split=
            True, data=tips)
```

## Barplot and Countplot

```python
# here 3 is the mean value of tip and that line is confidence index which is 95%
sns.catplot(x='sex', y='tip', kind='bar', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='bar', hue='sex',data=tips)
```

`COUNTPLOT work similar to histogram where histogram counts numerical value where countplot counts number of categorical value `

```python
sns.catplot(x="sex", kind='count', hue='smoker',data=tips)
```

# HeatMap

`Heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors`

```python
flight = sns.load_dataset('flights')
```

```python
flight.head(10)
```

```python
x = flight.pivot_table(index='year', columns='month', aggfunc='sum', values ='passengers')
```

```python
sns.heatmap(x) # heatmap with no of traveller each month in specific year
```
